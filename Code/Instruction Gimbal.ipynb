```python
#APPROVE 

q1, q2, q3 = sp.symbols('q1 q2 q3', real=True)

DH = sp.Matrix([
    [q1 + sp.pi / 2, 0, 0, sp.pi / 2],
    [q2 - sp.pi / 2, 0, 0, -sp.pi / 2],
    [q3, 0, 0, 0]
])

A01 = sp.Matrix([
    [sp.cos(DH[0, 0]), -sp.sin(DH[0, 0]) * sp.cos(DH[0, 3]), sp.sin(DH[0, 0]) * sp.sin(DH[0, 3]), 0],
    [sp.sin(DH[0, 0]), sp.cos(DH[0, 0]) * sp.cos(DH[0, 3]), -sp.cos(DH[0, 0]) * sp.sin(DH[0, 3]), 0],
    [0, sp.sin(DH[0, 3]), sp.cos(DH[0, 3]), 0],
    [0, 0, 0, 1]
])

A12 = sp.Matrix([
    [sp.cos(DH[1, 0]), -sp.sin(DH[1, 0]) * sp.cos(DH[1, 3]), sp.sin(DH[1, 0]) * sp.sin(DH[1, 3]), 0],
    [sp.sin(DH[1, 0]), sp.cos(DH[1, 0]) * sp.cos(DH[1, 3]), -sp.cos(DH[1, 0]) * sp.sin(DH[1, 3]), 0],
    [0, sp.sin(DH[1, 3]), sp.cos(DH[1, 3]), 0],
    [0, 0, 0, 1]
])

A23 = sp.Matrix([
    [sp.cos(DH[2, 0]), -sp.sin(DH[2, 0]) * sp.cos(DH[2, 3]), sp.sin(DH[2, 0]) * sp.sin(DH[2, 3]), 0],
    [sp.sin(DH[2, 0]), sp.cos(DH[2, 0]) * sp.cos(DH[2, 3]), -sp.cos(DH[2, 0]) * sp.sin(DH[2, 3]), 0],
    [0, sp.sin(DH[2, 3]), sp.cos(DH[2, 3]), 0],
    [0, 0, 0, 1]
])

T03 = A01 * A12 * A23

R01 = A01[0:3, 0:3]
R12 = A12[0:3, 0:3]
R23 = A23[0:3, 0:3]
# R01:This rotation matrix R01 is extracted from the transformation matrix A01. 
# It represents the orientation of frame 1 relative to frame 0. In DH parameter conventions, 
# each frame transformation consists of rotation and translation. The rotational component R01 
# embodies how frame 1 is rotated concerning frame 0.

#R12:Similarly, R12 is extracted from the transformation matrix A12. 
# It denotes the orientation of frame 2 relative to frame 1. 
# This matrix captures the rotation from frame 2 to frame 1.

#R23:R23 is extracted from the transformation matrix A23. It signifies the 
# orientation of frame 3 relative to frame 2. This matrix encapsulates the rotation from frame 3 to frame 2.

#Each of these 3x3 rotation matrices (R01, R12, R23) defines a rotation in three-dimensional 
# space about the x, y, and z axes (or their respective frame-specific axes). 
# These matrices contain trigonometric functions representing rotations around these axes.

#These matrices are crucial in robotics for understanding 
# how different parts of a robotic system are oriented concerning each other. 
# They're fundamental in transformations, allowing the computation of positions and orientations 
# of various components within a robotic system based on joint angles (q1, q2, q3 in this case) 
# or other kinematic parameters.

#review
print('Forward Kinematic = ',T03)
print('Rotation 01 = ',R01)
print('Rotation 12 = ',R12)
print('Rotation 23 = ',R23)
```

```python
#APPROVE

#Transalational Part of the jacobian in null

JP = sp.zeros(3, 3)

z0 = sp.Matrix([0,0,1])
z1 = R01 * z0
z2 = R01 * R12 * z0

Jw3 = sp.Matrix.hstack(z0, z1, z2)

JO = Jw3

J = sp.Matrix.vstack(JP, JO) #48

#review
print('JP',JP)
print('z0',z0)
print('z1',z1)
print('z2',z2)

print('Jw3', Jw3)
print('J',J)
```

```python
#APPROVE

# Link masses
m1, m2, m3 = sp.symbols('m1 m2 m3', real=True) 
#sp.symbols('m1 m2 m3', real=True): This line creates three symbolic variables m1, m2, and m3 using sympy.symbols. 
#The real=True argument specifies that these symbols are real-valued.


# Centers of masses
com1x, com1z = sp.symbols('com1x com1z', real=True)
com2x, com2z = sp.symbols('com2x com2z', real=True)
com3z = sp.symbols('com3z', real=True)

# Inertia tensor elements
Ixx1, Iyy1, Izz1, Ixz1 = sp.symbols('Ixx1 Iyy1 Izz1 Ixz1', real=True)

Il1 = sp.Matrix([
    [Ixx1, 0, -Ixz1],
    [0, Iyy1, 0],
    [-Ixz1, 0, Izz1]
])

Ixx2, Iyy2, Izz2, Ixz2 = sp.symbols('Ixx2 Iyy2 Izz2 Ixz2', real=True)

Il2 = sp.Matrix([
    [Ixx2, 0, -Ixz2],
    [0, Iyy2, 0],
    [-Ixz2, 0, Izz2]
])

Ixx3, Iyy3, Izz3 = sp.symbols('Ixx3 Iyy3 Izz3', real=True)

Il3 = sp.Matrix([
    [Ixx3, 0, 0],
    [0, Iyy3, 0],
    [0, 0, Izz3]
])


#reivew
print('Il1',Il1)
print('Il2',Il2)
print('Il3',Il3)
```

```python
#APPROVE

# Positions of the center of mass in the base frame 
# verified
pl1 = R01 * sp.Matrix([com1x, 0, com1z])
pl2 = R01 * R12 * sp.Matrix([com2x, 0, com2z])
pl3 = R01 * R12 * R23 * sp.Matrix([0, 0, com3z])



# Translational Jacobians
Jpl1 = sp.simplify(sp.Matrix([
    [z0.cross(pl1),
    sp.zeros(3,1),
    sp.zeros(3,1)]
])) # [(3x1) , (3x1) , (3x1)]

Jpl2 = sp.simplify(sp.Matrix([
    [z0.cross(pl1),
    z1.cross(pl2),
    sp.zeros(3,1)]
]))

Jpl3 = sp.simplify(sp.Matrix([
    [z0.cross(pl1),
    z1.cross(pl2),
    z2.cross(pl3)]
]))

# Rotational Jacobians
Jol1 = sp.Matrix([[
    z0,
    sp.zeros(3,1),
    sp.zeros(3,1)
]])

Jol2 = sp.Matrix([[
    z0,
    z1,
    sp.zeros(3,1)
]])

Jol3 = sp.Matrix([[
    z0,
    z1,
    z2
]])

#review
print('pl1', pl1)
print('pl2', pl2)
print('pl3', pl3)

print('Jpl1', Jpl1)
print('Jpl2', Jpl2)
print('Jpl3', Jpl3)

print('Jol1', Jol1)
print('Jol2', Jol2)
print('Jol3', Jol3)

print('--------------------Blank--------------------')

print ('z2 cross pl3', z2.cross(pl3))
print('New1',((m1 * (Jpl1.T * Jpl1)) + Jol1.T * R01 * Il1 * R01.T * Jol1))

```
